```python
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt
import math
```

```python
image = np.asarray(Image.open('800x825.jpg'))[:,:800]
plt.imshow(image)
```

```python
width, height = image.shape[:-1]

# Create a new empty image
new_width, new_height = 28, 28

scale = width / new_width

new_image = np.zeros((new_width, new_height, 3), np.uint8)

print(scale)

for col in range(0, width, int(scale)):
    for row in range(0, height, int(scale)):
        new_image[int(col//scale), int(row//scale)] = image[col, row]
        
new_image[:,-1] = new_image[:, -2]
        
plt.imshow(new_image)
```
