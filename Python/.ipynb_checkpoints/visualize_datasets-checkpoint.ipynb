```python
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
from random import choice
import os
from mpl_toolkits.axes_grid1 import ImageGrid
```

```python
ROOT = 'raw_datasets'

sample_images = []
sample_labels = []
dirs = list(os.walk(ROOT))

parent_dir = dirs[0][0]
second_level = dirs[8][0]
for i in range(1, len(dirs), 2):
    sample_labels.append(dirs[i][1][0])

for i in range(2, len(dirs), 2):
    img = Image.open(os.path.join(dirs[i][0], choice(dirs[i][2])))
    sample_images.append(np.asarray(img))
```

```python
plt.imshow(sample_images[1])
```

```python
fig = plt.figure(figsize=(10., 10.))
grid = ImageGrid(fig, 111,  # similar to subplot(111)
                 nrows_ncols=(2, 3),  # creates 2x2 grid of axes
                 axes_pad=0.5,  # pad between axes in inch.
                 share_all=True
                 )

for i, (ax, im) in enumerate(zip(grid, sample_images)):
    # Iterating over the grid returns the Axes.
    ax.imshow(im)
    ax.set_title(sample_labels[i])
    
grid[0].get_yaxis().set_ticks([])
grid[0].get_xaxis().set_ticks([])

fig.delaxes(grid[5])
plt.savefig(f'sample_images_{len(sample_labels)}.png', transparent=True)
plt.savefig(f'sample_images_{len(sample_labels)}_white.png', facecolor='white', transparent=False)
```
