```python
in_size = 240
out_size = 60

out_list = [0] * out_size*out_size*3

for i in range(0, in_size, 4):
    for j in range(0, in_size, 4):
        # Per pixel
        row = i * in_size*3
        max_r = 0
        max_g = 0
        max_b = 0
        for a in range(4):
            for b in range(4):
                pos = row + j * 3 + a*in_size*3 + b*3
                r = l[pos]
                g = l[pos + 1]
                b = l[pos + 2]
                if r > max_r:
                    max_r = r
                if g > max_g:
                    max_g = g
                if b > max_b:
                    max_b = b
                    
        out_list[i//4 * out_size*3 + j*3//4] = max_r
        out_list[i//4 * out_size*3 + 1 + j*3//4] = max_g
        out_list[i//4 * out_size*3 + 2 + j*3//4] = max_b
```

```python
out_array = np.asarray(out_list).reshape((out_size, out_size, 3))
#print(out_array)
plt.imshow(out_array)
```
